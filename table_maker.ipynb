```python
# import mods pandas, csv
import pandas as pd
from pathlib import Path
sset = "VOW"
```

```python
# path and read  master csv
file_one = Path("Resources/cards.csv")
```

```python
# make pandas df
df = pd.read_csv(file_one, dtype= str)
df.head(1)
```

```python
# create a dataframe of one card set
set_df = df.loc[df['setCode'] == sset]
# selected columns for inventory csv
set_df = set_df[['setCode',  'rarity', 'manaCost', 'colorIdentity', 'types', 'subtypes', 'power', 'toughness']]
#rename columns
set_df.rename(columns={"setCode": "set",  'manaCost': 'mana cost', })

# set_df = set_df.drop_duplicates(subset=['name']) # drop alernate prints
# find and replace values if necessary (find " " replace with 0)
set_df.fillna(0, inplace=True)
#re order columns
set_df = set_df[['setCode',  'rarity', 'manaCost', 'colorIdentity', 'types', 'subtypes', 'power', 'toughness' 
]]
set_df.head(3)
```

```python
# save as csv file to be used with postgressSQL db
set_df.to_csv(f'{sset}.csv', index=False)
```
