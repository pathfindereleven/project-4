```python
# Import required libraries
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Read the sales data
file_path = Path("resources\machine_num.csv")
df_sales = pd.read_csv(file_path)

# Display sample data
df_sales.head()
```

```python

df_sales = df_sales.drop(columns=['date'])
df_sales.head()
```

```python
num_df = pd.get_dummies(df_sales['setCode'])
num_df
```

```python
frames = [num_df, df_sales]
result = pd.concat(frames, axis =1)
result
```

```python
# add the total cards in set to results then feed results into meachine
def calculate_total_card(row):
    if row['KHM'] == 1:
        return 460
    elif row['MID'] == 1:
        return 480
    elif row['NEO'] == 1:
        return 574
    elif row['VOW'] == 1:
        return 505
    elif row['WAR'] == 1:
        return 312
    elif row['XLN'] == 1:
        return 299
    # Add other conditions as needed
    else:
        return None

# Add the 'total_card' column using the apply function
result['total_card'] = result.apply(calculate_total_card, axis=1)
result
```

```python
feed_df = result.drop(columns=['setCode'])
feed_df.head(3)
```

```python
y = feed_df['price']
X = feed_df.drop(columns=['price'])
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
# Train the data
classifier.fit(X_train, y_train)
```

```python
# Score the model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
df_percent = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
df_percent['Percentage'] = (df_percent['Actual'] / df_percent['Prediction'] * 100 -100).round(2) 
df_percent
```
