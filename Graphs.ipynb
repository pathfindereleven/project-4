```python
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# open csv to pandas df 
file_path = Path("prediction_m1.csv")
m1_predic = pd.read_csv(file_path)
m1_predic.head(3)

file_path = Path("prediction_m2.csv")
m2_predic = pd.read_csv(file_path)

m1_predic = m1_predic.rename(columns={'Predictions': 'Machine1'})
m1_predic['Machine2'] = m2_predic["Predictions"]
m1_predic.head(3)
```

```python
xln_df = m1_predic.loc[m1_predic['setCode'] == "XLN"]
lines = xln_df.plot.line(x='week_num', y=['Machine1', 'Machine2', 'price'])
```
