```python
#open predict csv and actual and combine
# open csv to pandas df 
file_path = Path("resources\m1VOW-pre.csv")
m1a_predic = pd.read_csv(file_path)

file_path = Path("resources\m2VOW-pre.csv")
m2a_predic = pd.read_csv(file_path)

file_path = Path("resources\m3VOW-pre.csv")
m3a_predic = pd.read_csv(file_path)


file_path = Path("resources\vow_actual.csv")
actual = pd.read_csv(file_path)

m1a_predic = m1a_predic.rename(columns={'Price Prediction': 'Machine1'})
m1a_predic['actual price'] = actual["price"]
m1a_predic['Machine2'] = m2a_predic ["Price Prediction"]
m1a_predic['Machine3'] = m3a_predic ["Price Prediction"]
m1a_predic.head(25)
```

```python
plt.plot(m1a_predic['week_num'], m1a_predic['Machine1'], color='blue', label='Machine 1')
plt.plot(m1a_predic['week_num'], m1a_predic['actual price'], color='red', label='actual')
plt.plot(m1a_predic['week_num'], m1a_predic['Machine2'], color='green', label='Machine 2')
plt.plot(m1a_predic['week_num'], m1a_predic['Machine3'], color='yellow', label='Machine 3')
plt.xlabel('Week Number')
plt.ylabel('Price')
plt.legend(loc="upper right")
plt.title('VOW Data')
plt.savefig('resources\VOW.png')
plt.show
```

```python
accuracy_df = m1a_predic[['week_num','Machine1', 'actual price']].copy()
accuracy_df.tail(3)
```

```python
accuracy_df['Percentage'] = (accuracy_df['actual price'] / accuracy_df['Machine1'] * 100 -100).round(2)
accuracy_df.tail(3)
```

```python

nonegs = []
for i in accuracy_df['Percentage']:
    if i >= 0:
        nonegs.append(i)
    if i < 0:
        i = i * -1
        nonegs.append(i)
accuracy_df['Machine1'] = nonegs
accuracy_df.head(10)
```

```python
avg=  accuracy_df['Machine1'].mean()
print(f'the Machine1 average error is {avg}')
```

```python
# calculate accuracy machine 2
accuracy_df['Machine2'] = m1a_predic['Machine2']
accuracy_df['Percentage2'] = (accuracy_df['actual price'] / accuracy_df['Machine2'] * 100 -100).round(2)

accuracy_df['Machine3'] = m1a_predic['Machine3']
accuracy_df['Percentage3'] = (accuracy_df['actual price'] / accuracy_df['Machine3'] * 100 -100).round(2)

accuracy_df.tail(3)
```

```python
nonegs = []
for i in accuracy_df['Percentage2']:
    if i >= 0:
        nonegs.append(i)
    if i < 0:
        i = i * -1
        nonegs.append(i)
accuracy_df['Machine2n-'] = nonegs
accuracy_df.tail(10)
```

```python
avg=  accuracy_df['Machine2n-'].mean()
print(f'the Machine2 average error is {avg}')
```

```python
# calculate average for machine 3
nonegs = []
for i in accuracy_df['Percentage3']:
    if i >= 0:
        nonegs.append(i)
    if i < 0:
        i = i * -1
        nonegs.append(i)
accuracy_df['Machine3n-'] = nonegs
accuracy_df.tail(25)
```

```python
avg=  accuracy_df['Machine3n-'].mean()
print(f'the Machine3 average error is {avg}')
```

```python
species = ("Machine1", "Machine2", "Machine3")
error = {
    'Training error': (4.64, 4.35, 4.04),
    'Prediction error': (10.67, 8.82, 10.07),
    
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in error.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percent')
ax.set_title('Percent Error by data set')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 12)
plt.savefig('resources\_bar.png')
plt.show()
```
